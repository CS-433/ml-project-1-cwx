```python
"""Load data"""
train_path = "data/train.csv/train.csv"
test_path = "data/test.csv/test.csv"
train_target, train_data, _ = load_csv_data(train_path)
_, test_data, ids = load_csv_data(test_path)
```

```python
train=[]
test=[]
loss=[]
types=[]
```

```python
train_target[train_target == -1] = 0
train_target = train_target.reshape((-1, 1))
train_data, test_data = fix_empty(train_data, test_data)
train_data, test_data = lognormal(train_data, test_data)
```

```python
model = Sequential(Linear(train_data.shape[1], 128), ReLU(), Linear(128, 128), ReLU(),
                   Linear(128, 2))
a,b,c=train_model(train_data[:200000], train_target[:200000], model, crit='cel', print_res=True, nb_epochs=500,
            mini_batch_size=100, lambda_l2=1e-5, learning_rate=1e-1, test_target=train_target[200000:],
            test_data=train_data[200000:],cosine=True)

loss.append(a)
train.append(b)
test.append(c)
types.append('With CEloss')
```

```python
train_target[train_target == 0] = -1
train_target = convert_to_one_hot(train_target)
"""Create dnn model and train the model"""

model = Sequential(Linear(train_data.shape[1], 128), ReLU(), Linear(128, 128), ReLU(),
                   Linear(128, 2))
a,b,c=train_model(train_data[:200000], train_target[:200000], model, crit='mse', print_res=True, nb_epochs=500,
            mini_batch_size=100, lambda_l2=1e-5, learning_rate=1e-1, test_target=train_target[200000:],
            test_data=train_data[200000:], cosine=True)
loss.append(a)
train.append(b)
test.append(c)
types.append('With MSEloss')
```

```python
plot_res(train,types,'Train accuracy')
plot_res(test,types,'Test accuracy')
```
